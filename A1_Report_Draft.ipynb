## **INTRODUCTION**

'Facial recognition is a biometric software application capable of uniquely identifying or verifying a person by comparing and analysing patterns based on the person's facial contours' (Techopedia 2019). Face recognition has recently become a very interesting research area and is regarded as an emerging technology in the image processing and recognition industry. Facial recognition is mainly used for identification and verification, but its application has expanded widely.
  
Numerous methods have been used for face recognition, but it still faces a lot of challenges. There are many factors that might affect the recognition technique such as variation in light, facial expression, pose etc.

This report presents a literature review on "Eigenfaces vs. Fisherfaces: Recognition Using Class Specific Linear Projection" by Belhumeur, Hespanha & Kriegman (1997). The article will be evaluated with regards to its content, innovation, technical quality, application, x-factor and presentation.

## **CONTENT**
  
The paper outlined a new approach for face recognition algorithm, which is insensitive to large variation in lighting direction and facial expression. This method is called Fisherface, derived from Fisher’s Linear Discriminant (FLD). Fisher’s Linear Discriminant is an example of a class-specific method where it tries to shape the scatter to make it reliable for classification, maximising the ratio of between-class scatter to that of within-class scatter.
  
Other methods of face recognition that have become quite prevalent namely Correlation, Eigenfaces, Linear Subspaces were compared with the method developed by the authors called Fisherface. The main problem of the experiment was to identify each person in the test images given a set of face images labelled with a person’s identity and an unlabelled set of face images from the same group. The experiments were done using a subset from the Harvard database and a database at Yale.
  
The authors performed extrapolation and interpolation experiments using the Harvard database, where lighting was systematically varied. All methods performed perfectly well when lighting is nearly frontal. When lighting was moved off the axis, the performance of Eigenface method was equivalent to correlation. Eigenface’s performance improved when the first three principal components were removed. Linear Subspace had error rates competitive with Fisherface method, but it required more storage and took three times as long. Fisherface method had lower error rates than the Eigenface method and required less computation time.
  
Tests with variation in facial expression, eyewear and lighting were also completed using the Yale database. Error rates were determined by the “leaving-one-out” strategy and recognition was performed using the nearest neighbour classifier. Fisherface method had error rates that were better than half that of any method as it performed well over lighting, facial expression and presence of glasses.

In tests on both Harvard and database, it can be concluded that the Fisherface method had lower error rates than Correlation, Eigenface method and Linear Subspace. However, there was no claim about the performance on larger databases. The authors also pointed out that there was no attempt to deal with pose variation and they assumed that the face has been located and aligned within the image.

## **INNOVATION**

Organisations today are motivated to improve facial recognition methods. Researchers are trying to find ways and solutions that can recognise an image in all situations encountered by users or businesses. However, there are many factors to consider like the variation in lighting, facial expression and angle of the face. The facial recognition system results are significantly affected, whether it be a small or big change in lighting, expression and pose which makes it harder for researchers to develop the perfect solution for various situations. 
 
The novelty presented in the paper is the authors’ new face algorithm, which is insensitive to large variation in lighting direction and facial expression. This new approach for face recognition is called Fisherface, a derivative of Fisher’s Linear Discriminant (FLD). FLD achieves greater between-class scatter, which results in simpler classification. In face recognition, there is a difficulty that the within-class matrix is singular. In order to overcome this complication, Fisherface method projects the image set to lower dimensional space such that the resulting within-class matrix is not singular. Hence, the Fisherface method maximises the ratio of the between-class scatter and within-class scatter, resulting to lower error rates in face recognition.
 
The Fisherface method developed by the authors of the article was compared against the existing face recognition techniques namely Correlation, Eigenfaces and Linear Subspace in this article. Advantages and disadvantages of each method were discussed and experimental tests with variation in lighting and facial expression were performed on the Harvard and Yale database to compare the different face recognition methods. Results on both tests showed that the Fisherface method performed better compared to the other popular face recognition methods.
